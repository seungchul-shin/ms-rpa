# Task
상위 20개 키워드 변화를 보기 위하여 3개년별로 키워드를 종합하여 분석

## Solution
다음의 Split-Apply-Combine 방식을 따른다.

* Step 1. 데이터프레임을 기준년도 2개를 조건으로하여 자료를 재분배
* Step 2. `collections.Counter` 객체를 활용하여 3개년마다 키워드를 카운트 및 정렬(단, 키워드만 추출할 것)
* Step 3. 상위 100개 키워드만 추출 후 `pd.DataFrame`으로 재구조화 및 저장

```python
ls = {}
COUNT = {}

for tpl in [(2011, 2013), (2014, 2016), (2017, 2019), (2020, 2021)]:
    ls[tpl] = df.query('연도 >= @tpl[0] and 연도 <= @tpl[1]').iloc[:, 1:].values.ravel()
    COUNT[tpl] = df.query('연도 >= @tpl[0] and 연도 <= @tpl[1]').shape[0]
```

```python
len(ls)
```

```python
COUNT
```

```python
counter_dc = {}

for year, content in ls.items():
    words = content[pd.notnull(content)]
    words_counter = Counter(words)
    counter_dc[year] = np.array(words_counter.most_common(100))[:, 0]
```

```python
counter_dc.get((2014, 2016))
```

```python
pd.DataFrame(counter_dc).to_csv("k3y_counter.csv", encoding='utf-8', index=False)
```
